```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv('103年汽車貨運調查原始資料.csv',encoding='big5')
b = pd.read_csv('104年汽車貨運調查原始資料.csv',encoding='big5')
c = pd.read_csv('105年汽車貨運調查-原始資料.csv',encoding='big5')
d = pd.read_csv('106年汽車貨運調查原始資料.csv',encoding='big5')
e = pd.read_csv('107年汽車貨運調查原始資料.csv',encoding='big5')
```

```python
raw = pd.concat([a,b,c,d,e])
raw.head()
```

```python
raw['地區別'].replace('N','北部地區',inplace = True)
raw['地區別'].replace('C','中部地區',inplace = True)
raw['地區別'].replace('S','南部地區',inplace = True)
raw['地區別'].replace('E','東部地區',inplace = True)
```

```python
x = pd.read_csv('def - Sheet1.csv',encoding='utf-8')
```

```python
for idx, row in x.iterrows():
  raw['商品編號'].replace(idx,x.iloc[idx,1],inplace = True)
raw
```

```python
raw['商品編號'].replace(201,'貨櫃貨',inplace = True)
raw['商品編號'].replace(200,'空貨櫃',inplace = True)
```

```python
y = pd.read_csv('def - Sheet2.csv',encoding='utf-8')
```

```python
for idx, row in y.iterrows():
  raw['裝貨地點'].replace(y.iloc[idx,0],y.iloc[idx,1],inplace = True)
  raw['卸貨地點'].replace(y.iloc[idx,0],y.iloc[idx,1],inplace = True)
raw
```

```python
raw['商品來源別'].replace(1,'進口貨',inplace = True)
raw['商品來源別'].replace(2,'出口貨',inplace = True)
raw['商品來源別'].replace(3,'轉口貨',inplace = True)
raw['商品來源別'].replace(4,'國產內銷貨',inplace = True)
```

```python
raw['低溫商品比率'].replace(1,'0%',inplace = True)
raw['低溫商品比率'].replace(2,'1%~未滿5%',inplace = True)
raw['低溫商品比率'].replace(3,'5%~未滿10%',inplace = True)
raw['低溫商品比率'].replace(4,'10%~未滿15%',inplace = True)
raw['低溫商品比率'].replace(5,'15%~未滿20%',inplace = True)
raw['低溫商品比率'].replace(6,'20%~未滿40%',inplace = True)
raw['低溫商品比率'].replace(7,'40%~未滿60%',inplace = True)
raw['低溫商品比率'].replace(8,'60%~未滿80%',inplace = True)
raw['低溫商品比率'].replace(9,'80%~未滿100%',inplace = True)
raw['低溫商品比率'].replace(0,'100%',inplace = True)
```

```python
r = raw.dropna()
r
```

```python
TC_pickup = r[r['裝貨地點'].str.contains('臺中') == True]
TC_destination = r[r['卸貨地點'].str.contains('臺中') == True]

TY_pickup = r[r['裝貨地點'].str.contains('桃園') == True]
TY_destination = r[r['卸貨地點'].str.contains('桃園') == True]

KH_pickup = r[r['裝貨地點'].str.contains('高雄') == True]
KH_destination = r[r['卸貨地點'].str.contains('高雄') == True]
```

```python
TC_pickup_commodity = TC_pickup.groupby('商品編號').sum()
TC_destination_commodity = TC_destination.groupby('商品編號').sum()

TY_pickup_commodity = TY_pickup.groupby('商品編號').sum()
TY_destination_commodity = TY_destination.groupby('商品編號').sum()

KH_pickup_commodity = KH_pickup.groupby('商品編號').sum()
KH_destination_commodity = KH_destination.groupby('商品編號').sum()
```

```python
TC_pickup_commodity.to_csv('TC_pickup_commodity.csv', index=True)
TC_destination_commodity.to_csv('TC_destination_commodity.csv', index=True)

TY_pickup_commodity.to_csv('TY_pickup_commodity.csv', index=True)
TY_destination_commodity.to_csv('TY_destination_commodity.csv', index=True)

KH_pickup_commodity.to_csv('KH_pickup_commodity.csv', index=True)
KH_destination_commodity.to_csv('KH_destination_commodity.csv', index=True)
```

```python
#######
TW_logistic_trend = r.groupby('調查年度').sum()
TW_logistic_trend.to_csv('TW_logistic_trend.csv',index=True)
```

```python
#######
TW_intracity_logistic_trend = r[r['裝貨地點'] == r['卸貨地點']].groupby('調查年度').sum()
TW_intracity_logistic_trend.to_csv('TW_intracity_logistic_trend.csv',index=True)
```

```python
TC_frames = [TC_pickup,TC_destination]
TC_orders = pd.concat(TC_frames)
TC_orders_count = TC_orders.groupby('調查年度').sum()

TY_frames = [TY_pickup,TY_destination]
TY_orders = pd.concat(TY_frames)
TY_orders_count = TY_orders.groupby('調查年度').sum()

KH_frames = [KH_pickup,KH_destination]
KH_orders = pd.concat(KH_frames)
KH_orders_count = KH_orders.groupby('調查年度').sum()
```

```python
all_frames = [TC_orders_count, TY_orders_count, KH_orders_count]
tier1_cities_order_count = pd.concat(all_frames)
tier1_cities_order_count.to_csv('tier1_cities_order_count.csv',index=False)
```
